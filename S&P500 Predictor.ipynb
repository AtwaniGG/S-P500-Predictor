```python
import yfinance as yf 
import pandas as pd
```

```python
sp500 = yf.Ticker("^GSPC")
```

```python
sp500 = sp500.history(period="max")
```

```python
sp500
```

```python
sp500.index
```

```python
sp500.plot.line(y="Close", use_index=True)
```

```python
del sp500["Dividends"]
del sp500["Stock Splits"]
```

```python
sp500["Tomorrow"] = sp500["Close"].shift(-1)
```

```python
sp500
```

```python
sp500["Target"] = (sp500["Tomorrow"] > sp500["Close"]).astype(int)

#Target price is when Tomorrow's price is greater than today's price 
# 1 if True
# 0 if False
```

```python
sp500 = sp500.loc["1995-01-01":].copy()
#shows dates after 01-01-1995
```

```python
sp500
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, min_samples_split= 100, random_state= 1)
train = sp500.iloc [:-100]
test = sp500.iloc [-100:]

predictors = ["Close", "Volume", "Open", "High", "Low"]
model.fit(train[predictors], train["Target"])
#n_estimators is the number of individual decision trees that will be trained
#min_samples_split to prevent tree from spliting and overfitting
#random_state = 1 allows you to run the model more than once and get similiar results to compare
```

```python
from sklearn.metrics import precision_score
preds = model.predict(test[predictors])
#precision_score is an accuracy metric
```

```python
preds = pd.Series(preds, index=test.index)
```

```python
precision_score(test["Target"], preds)
#means that the stock went up (precision_score) of times 
```

```python
combined = pd.concat([test["Target"], preds], axis = 1)

#axis = 1 means treat each of the inputs as a seperate column
```

```python
combined.plot()
```

```python
#backtesting
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
def backtest(data, model, predictors, start= 2500, step=250):  #
    all_predictions = []
    
    for i in range(start, data.shape[0],step):
        train = data.iloc[0:i].copy() #all years prior to current year
        test = data.iloc[i:(i+step)].copy() #current year
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)

    return pd.concat(all_predictions)   #takes all data frames and combines them into 1 dataframe
```

```python
predictions = backtest(sp500, model, predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
predictions["Target"].value_counts() / predictions.shape[0]
```

```python
#calculatiing mean closing price in last 2 days, 5 days (trading week), 60 days(3 trading months ), 250 days (trading year), 1000 days (4 trading years)
#then compare the ratio of today's closing price with the output 
horizons = [2,5,60,250,1000]

new_predictors = []

for horizon in  horizons:
    rolling_averages = sp500.rolling(horizon).mean()
    ratio_column = f"Close Ratio_{horizon}"
    sp500[ratio_column] = sp500["Close"] / rolling_averages["Close"]

    trend_column = f"Trend_{horizon}"
    sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()["Target"]

    new_predictors += [ratio_column, trend_column]
```

```python
sp500 = sp500.dropna()
```

```python
sp500
```

```python
model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors])[:,1]
    preds[preds >=.6] = 1
    preds[preds <.6] = 0
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
predictions = backtest(sp500, model, new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
#check precision score with this

precision_score(predictions["Target"], predictions["Predictions"] )
```
